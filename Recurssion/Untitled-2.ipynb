```python
# Bubble sort

def bubble_sort(arr):
    n = len(arr)
    for turn in range(n):
        for j in range(n-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr

print(bubble_sort([5, 3, 8, 6, 7, 2]))

# Selection sort

def selection_sort(arr):
    n = len(arr)
    for turn in range(n):
        min_pos = turn
        for j in range(turn+1, n):
            if arr[j] < arr[min_pos]:
                min_pos = j
        arr[turn], arr[min_pos] = arr[min_pos], arr[turn]

    return arr

print(selection_sort([5, 3, 8, 6, 7, 2]))

# Insertion sort

def insertion_sort(arr):
    n = len(arr)
    for turn in range(1, n):
        j = turn
        while j > 0 and arr[j] < arr[j-1]:
            arr[j], arr[j-1] = arr[j-1], arr[j]
            j -= 1
        
    return arr

print(insertion_sort([5, 3, 8, 6, 7, 2]))

# Counting sort (only for positive integers)

def counting_sort(arr):
    n= len(arr)
    max_val = max(arr)
    count = [0] * (max_val+1)

    for i in range(n):
        count[arr[i]] += 1
    
    k = 0
    for i in range(max_val+1):
        for j in range(count[i]):
            arr[k] = i
            k += 1
    
    return arr


counting_sort([5, 3, 8, 6, 7, 2])
```

```python
#  Rotated sorted array:

def rotated_sorted_array(arr, target, si, ei):
    if si > ei:
        return -1
    mid = si + (ei - si) // 2

    if arr[mid] == target:
        return mid
    
    if arr[si] <= arr[mid]:
        if target >= arr[si] and target <= arr[mid]:
            return rotated_sorted_array(arr, target, si, mid-1)
        return rotated_sorted_array(arr, target, mid+1, ei)
    
    if target >= arr[mid] and target <= arr[ei]:
        return rotated_sorted_array(arr, target, mid+1, ei)
    return rotated_sorted_array(arr, target, si, mid-1)

print(rotated_sorted_array([4,5,6,7,0,1,2], 0, 0, 6))

# TC: O(nlogn)
```

```python
# implementing stack using array:

class Stack:
    def __init__(self):
        self.arr = []

    def push(self, x):
        self.arr.append(x)
    
    def pop(self):
        if len(self.arr) == 0:
            return -1
        return self.arr.pop()
    
    def peek(self):
        if len(self.arr) == 0:
            return -1
        return self.arr[-1]
    
    def size(self):
        return len(self.arr)
    
    def isEmpty(self):
        return len(self.arr) == 0
    
    def printStack(self):
        while len(self.arr) != 0:
            print(self.arr.pop())

    def pushatBottom(self, x):
        if len(self.arr) == 0:
            self.arr.append(x)
            return
        temp = self.arr.pop()
        self.pushatBottom(x)
        self.arr.append(temp)

    def reverse(self):
        if len(self.arr) == 0:
            return
        temp = self.arr.pop()
        self.reverse()
        self.pushatBottom(temp)
        
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.printStack()
```

```python
#  Implementing stack using linked list:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class StackLL:
    def __init__(self):
        self.head = None

    def push(self, x):
        if self.head == None:
            self.head = Node(x)
            return
        new_node = Node(x)
        new_node.next = self.head
        self.head = new_node

    def pop(self):
        if self.head == None:
            return -1
        temp = self.head
        self.head = self.head.next
        return temp.data
    
    def peek(self):
        if self.head == None:
            return -1
        return self.head.data
    
    def size(self):
        temp = self.head
        count = 0
        while temp != None:
            count += 1
            temp = temp.next
        return count
    
    def isEmpty(self):
        return self.head == None
    
    def printStack(self):
        temp = self.head
        while temp != None:
            print(temp.data)
            temp = temp.next

    def reverse(self):
        prev = None
        current = self.head
        while current != None:
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev

    

s = StackLL()
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.pop()
s.printStack()

        
```

```python
# Reverse the linked list:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def push(self, x):
        if self.head == None:
            self.head = Node(x)
            return
        new_node = Node(x)
        new_node.next = self.head
        self.head = new_node
    
    def printList(self):
        temp = self.head
        while temp != None:
            print(temp.data)
            temp = temp.next
    
    def reverse(self):
        prev = None
        curr = self.head
        while curr != None:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        self.head = prev

    def isCycle(self):
        slow = self.head
        fast = self.head
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast: return True
        return False
    
    def removeCycle(self):
        slow = self.head
        fast = self.head
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast: break
        slow = self.head
        while slow.next != fast.next:
            slow = slow.next
            fast = fast.next
        fast.next = None
    
    def getMid(self):
        slow = self.head
        fast = self.head.next
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next
        return slow
    
    def merge(self, l1, l2):
        dummy = Node(0)
        temp = dummy
        while l1 != None and l2 != None:
            if l1.data < l2.data:
                temp.next = l1
                l1 = l1.next
            else:
                temp.next = l2
                l2 = l2.next
            temp = temp.next
        
        while l1 != None:
            temp.next = l1
            l1 = l1.next
            temp = temp.next
        
        while l2 != None:
            temp.next = l2
            l2 = l2.next
            temp = temp.next
        return dummy.next
    
    def mergeSort(self, head):
        if head == None or head.next == None:
            return head
        
        mid = self.getMid()
        nextmid = mid.next
        mid.next = None
        left = self.mergeSort(head)
        right = self.mergeSort(nextmid)
        return self.merge(left, right)

    def zigzag(self):
        mid = self.getMid()
        prev = None
        curr = mid.next
        mid.next = None
        while curr != None:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        
        rightHead = prev
        leftHead = self.head
        nextleft = nextRight = None

        while leftHead != None and rightHead != None:
            nextleft = leftHead.next
            leftHead.next = rightHead
            leftHead = nextleft
            nextRight = rightHead.next
            rightHead.next = leftHead
            rightHead = nextRight
        return leftHead

    def deleteKey(self, key):
        temp = self.head
        prev = None

        while temp != None:
            if temp.data == key:
                if prev:
                    prev.next = temp.next
                else:
                    self.head = temp.next
                temp = None
            prev = temp
            temp = temp.next
        return self.head
    
    def removeNthnodeFromEnd(self, n):
        # size of the linked list
        temp = self.head
        size = 0
        while temp != None:
            size += 1
            temp = temp.next
        if n > size:
            return self.head
        if n == size:
            return self.head.next
        temp = self.head

        for i in range(size - n - 1):
            temp = temp.next
        temp.next = temp.next.next
        return self.head


```

```python
# using the inbuilt queue implementing reverse the first k elements:

from queue import Queue

def reverse_k_elements(q, k):
    stack = []
    for i in range(k):
        stack.append(q.get())
    while len(stack) != 0:
        q.put(stack.pop())
    for i in range(q.qsize() - k):
        q.put(q.get())
    return q

    #TC: O(n)
    #SC: O(n)

#call the function:

q = Queue()
q.put(1)
q.put(2)
q.put(3)

print(reverse_k_elements(q, 2).queue)
```

```python
# interleave the first half of the queue with the second half:

from queue import Queue

def interleave(q):
    q1 = Queue()
    half = q.qsize() // 2
    for i in range(half):
        q1.put(q.get())
    while not q1.empty():
        q.put(q1.get())
        q.put(q.get())
    return q

q = Queue()
q.put(1)
q.put(2)
q.put(3)
q.put(4)
print(interleave(q).queue)


#TC: O(n)
#SC: O(n)
```

```python
# Fractional Knapsack

# steps to do
# 1. calculate the ratio of value and weight
# 2. sort the items based on the ratio
# 3. initialize the result and capacity of the knapsack
# 4. iterate through the items and add the items to the knapsack
# 5. if adding the item doesn't exceed the capacity, add the item to the result
# 6. if adding the item exceeds the capacity, add the fraction of the item to the result

def fractional_knapsack(value, weight, capacity):

    index = list(range(len(value)))
    ratio = [v/w for v,w in zip(value, weight)]
    index.sort(key = lambda x: ratio[x], reverse=True)
    max_value = 0
    fractions = [0]*len(value) # not necessary this is only to keep the track of the fractions of part we are taking in the bag
    # print(fractions)
    for i in index:
        if weight[i] <= capacity:
            max_value += value[i]
            capacity -= weight[i]
            fractions[i] = 1
        else:
            fractions[i] = capacity/weight[i]
            max_value += value[i]*fractions[i]
            break
    return max_value, fractions     

print(fractional_knapsack([60,100,120], [10,20,30], 50))
```

```python
# Minim Absolute difference sum of two arrays

# logic -- if we do the abs of the values which are close to each other then their difference will be minimum -- so we can sort bothe the arrays in the ascending order and then we can find the minimum difference between the two arrays

def min_abs_diff_sum(arr1, arr2):
    arr1.sort()
    arr2.sort()
    minimum_diff = 0
    for i in range(len(arr1)):
        minimum_diff += abs(arr1[i] - arr2[i])
    
    return minimum_diff

print(min_abs_diff_sum([4,1,8,7], [2,3,6,5]))

# TC: O(nlogn)
```

```python
# Indian Coin Change Problem 

# logic -- we can sort the coins in the descending order and then we can find the minimum number of coins required to make the change

def coin_change(coins, amount):
    coins.sort(reverse=True)
    count = 0
    for i in range(len(coins)):
        if coins[i] <= amount:
            while coins[i] <= amount:
                amount -= coins[i]
                count += 1

    return count

print(coin_change([1,2,5,10,20,50,100,500,1000], 93))

# TC: O(n)

# More optimized approach

def coin_change_optimized(coins, amount):
    coins.sort(reverse=True)
    count = 0
    for i in range(len(coins)):
        if coins[i] <= amount:
            count += amount // coins[i]   # this will give the number of coins required
            amount = amount % coins[i]   # this will give the remaining amount after the coins are used
    return count

print(coin_change_optimized([1,2,5,10,20,50,100,500,1000], 93))

# dry run
# 93 - 50 = 43
# 43 - 20 = 23
# 23 - 20 = 3
# 3 - 2 = 1
# 1 - 1 = 0

# TC: O(n)

```

```python
# Chocola Problem

# logic -- we can sort the chocolates cost for vertical cost and the horizontal cost in the descending order and then we can find the minimum cost to break the choclates

def chocola_problem(horizontal_cost, vertical_cost):
    horizontal_cost.sort(reverse=True)
    vertical_cost.sort(reverse=True)
    h_pointer, v_pointer = 0, 0
    h_pieces, v_pieces = 1, 1
    cost = 0

    while h_pointer < len(horizontal_cost) and v_pointer < len(vertical_cost):
        if horizontal_cost[h_pointer] > vertical_cost[v_pointer]:
            h_pieces += 1
            cost += horizontal_cost[h_pointer] * v_pieces
            h_pointer += 1
        else:
            v_pieces += 1
            cost += vertical_cost[v_pointer] * h_pieces
            v_pointer += 1
    while h_pointer < len(horizontal_cost):
        cost += horizontal_cost[h_pointer] * v_pieces
        h_pointer += 1

    while v_pointer < len(vertical_cost):
        cost += vertical_cost[v_pointer] * h_pieces
        v_pointer += 1

    return cost

print(chocola_problem([2,1,3,1,4], [4,1,2]))
```

```python
# Diameter of tree with O(n) TC

class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    

class Info:
    def __init__(self, diameter, height):
        self.diameter = diameter
        self.height = height

def diameter_of_tree(root):
    if not root:
        return Info(0,0)
    
    left_info = diameter_of_tree(root.left)
    right_info = diameter_of_tree(root.right)

    diameter = max(left_info.height + right_info.height + 1, max(left_info.diameter, right_info.diameter))
    height = 1 + max(left_info.height, right_info.height)

    return Info(diameter, height)

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print(diameter_of_tree(root).diameter)
```

```python
# Top View of tree:

from collections import deque
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

def top_view(root):
    if not root: return
    q = deque()
    mapp = {}

    q.append((root,0))

    while q:
        node, hd = q.popleft()
        if hd not in mapp:
            mapp[hd] = node.data
        if node.left:
            q.append((node.left, hd-1))
        if node.right:
            q.append((node.right, hd+1))

    for i in sorted(mapp):
        print(mapp[i], end=" ")      


def bottom_view(root):
    if not root: return
    q = deque()
    mapp = {}

    q.append((root,0))

    while q:
        node, hd = q.popleft()
        mapp[hd] = node.data
        if node.left:
            q.append((node.left, hd-1))
        if node.right:
            q.append((node.right, hd+1))

    for i in sorted(mapp):
        print(mapp[i], end=" ")       

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)

#    1
#   / \
#  2   3
# / \ / \
#4   5   6

top_view(root)
print()
bottom_view(root)
```

```python
# Tree has univalued nodes
from collections import deque
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def is_univalued(root):
    if not root: return True
    q = deque()
    q.append(root)
    mapp = {}

    while q:
        node = q.popleft()
        if node.val in mapp:
            return False
        mapp[node.val] = 1
        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)

    return mapp[root.val] == 1

root = Node(1)
root.left = Node(1)
root.right = Node(3)


print(is_univalued(root))

```

```python
# Maximum Sum in BST -- Given a BST -- find the maximum sum of a node of a subtree which is also a BST


class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class Info:
    def __init__(self, is_bst, size, sum, min, max):
        self.is_bst = is_bst
        self.size = size
        self.sum = sum
        self.min = min
        self.max = max

def max_sum_in_bst(root):
    if not root: return Info(True, 0,0, float('inf'), float('-inf'))

    left = max_sum_in_bst(root.left)
    right = max_sum_in_bst(root.right)

    size = left.size + right.size + 1
    sum = left.sum + right.sum + root.val
    min_val = min(root.val, min(left.min, right.min))
    max_val = max(root.val, max(left.max, right.max))

    if left.is_bst and right.is_bst and left.max < root.val and right.min > root.val:
        return Info(True, size, sum, min_val, max_val)
    
    return Info(False, max(left.size, right.size), max(left.sum, right.sum), 0, 0)

root = Node(5)
root.left = Node(9)
root.right = Node(2)
root.left.left = Node(6)
root.left.left.left = Node(8)
root.left.left.right = Node(7)
root.right.right = Node(3)

max_sum_in_bst(root).sum
```

```python
# Heap Sort

# Given an array -- sort the array using heap sort

arr = [10,20,15,12,40,25,18]

def heap_sort(arr):
    # creat max heap for the given array to sort the array in ascending order
    n = len(arr)


    # Last non-leaf node of the tree
    for i in range(n//2 -1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)
    
    return arr


def heapify(arr, n, i):
    left = 2 * i + 1
    right = 2 * i + 2
    largest = i

    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right
    
    if largest != i:
        arr[i],arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
    
    return arr


heap_sort(arr)

```

```python
# Near by Cars

# Given a list of cars and their positions -- find the nearest car to the given car from the origin

# Approach 1 -- Using Heap

from heapq import heappush, heappop

def nearest_cars(cars, k):
    heap = []
    for i in range(len(cars)):
        heappush(heap, (cars[i][0] **2 + cars[i][1] **2, i))

    res = []
    for i in range(k):
        res.append(cars[heappop(heap)[1]])

    return res

cars = [[3,4],[5,6],[7,8], [1,2]]
k = 2

nearest_cars(cars, k)


# Dry Run

# cars = [[3,4],[5,6],[7,8], [1,2]]
# k = 2

# iteration 1:
# heappush(heap, (3**2 + 4**2, 0)) => heappush(heap, (25, 0))
# heappush(heap, (5**2 + 6**2, 1)) => heappush(heap, (61, 1))
# heappush(heap, (7**2 + 8**2, 2)) => heappush(heap, (113, 2))
# heappush(heap, (1**2 + 2**2, 3)) => heappush(heap, (5, 3))

# heap = [(5, 3), (25, 0), (61, 1), (113, 2)]

# iteration 2:

# res = []
# res.append(cars[heappop(heap)[1]) => res.append(cars[3]) => res.append([1,2])
# res.append(cars[heappop(heap)[1]) => res.append(cars[0]) => res.append([3,4])

# res = [[1,2],[3,4]]

# Time Complexity: O(nlogn) -- for heapify
# Space Complexity: O(n) -- for heap

# Approach 2 -- Using Sorting

def nearest_cars(cars, k):
    cars.sort(key = lambda x: x[0] **2 + x[1] **2)
    return cars[:k]

cars = [[3,4],[5,6],[7,8], [1,2]]
k = 2

nearest_cars(cars, k)

# TC: O(nlogn) -- for sorting
# SC: O(1) -- no extra space used
```

```python
# Connect N ropes with minimum cost

# Given n ropes of different lengths, we need to connect these ropes into one rope. We can connect only 2 ropes at a time. The cost required to connect 2 ropes is equal to the sum of their lengths. The length of this connected rope is also equal to the sum of their lengths. This process is repeated until n ropes are connected into a single rope. Find the minimum cost to connect all ropes.

# Approach 1 -- Using Heap

from heapq import heappush, heappop

def connect_ropes(ropes):
    heap = []

    for i in range(len(ropes)):
        heappush(heap, ropes[i])

    res = 0

    while len(heap)> 1:
        first = heappop(heap)
        second = heappop(heap)

        res += first + second
        heappush(heap, first + second)

    return res

ropes = [4,3,2,6]
connect_ropes(ropes)

# Dry Run

# ropes = [4,3,2,6]

# iteration 1:
# heappush(heap, 4)
# heappush(heap, 3)
# heappush(heap, 2)
# heappush(heap, 6)

# heap = [2,3,4,6]

# iteration 2:
# first = heappop(heap) => first = 2
# second = heappop(heap) => second = 3

# res += first + second => res += 2 + 3 => res = 5

# heappush(heap, first + second) => heappush(heap, 5)

# heap = [4,5,6]

# iteration 3:
# first = heappop(heap) => first = 4
# second = heappop(heap) => second = 5

# res += first + second => res += 4 + 5 => res = 9

# heappush(heap, first + second) => heappush(heap, 9)

# heap = [6,9]

# iteration 4:

# first = heappop(heap) => first = 6
# second = heappop(heap) => second =
# res += first + second => res += 6 + 9 => res = 15
# heappush(heap, first + second) => heappush(heap, 15)

# heap = [15]

# Time Complexity: O(nlogn) -- for heapify
# Space Complexity: O(n) -- for heap

# Approach 2 -- Using Sorting

def connect_ropes(ropes):
    ropes.sort()
    res = 0

    for i in range(1, len(ropes)):
        ropes[i] = ropes[i] + ropes[i-1]
        res += ropes[i]

    return res

ropes = [4,3,2,6]
connect_ropes(ropes)

# Dry Run

# ropes = [4,3,2,6]
# ropes.sort() => ropes = [2,3,4,6]

# iteration 1:
# ropes[1] = ropes[1] + ropes[0] => ropes[1] = 3 + 2 => ropes[1] = 5
# res += ropes[1] => res += 5 => res = 5

# iteration 2:
# ropes[2] = ropes[2] + ropes[1] => ropes[2] = 4 + 5 => ropes[2] = 9
# res += ropes[2] => res += 9 => res = 14

# iteration 3:
# ropes[3] = ropes[3] + ropes[2] => ropes[3] = 6 + 9 => ropes[3] = 15

# res += ropes[3] => res += 15 => res = 29

# Time Complexity: O(nlogn) -- for sorting
# Space Complexity: O(1) -- no extra space used
```

```python
# Sliding Window Maximum -- Using heap

from heapq import heappush, heappop

def sliding_window_maximum(arr, k):
    heap = []
    res = [] 

    for i in range(k):
        heappush(heap, (-arr[i], i))

    res.append(-heap[0][0])

    for i in range(k, len(arr)):
        while heap and heap[0][1] <= i - k:
            heappop(heap)
        heappush(heap, (-arr[i], i))
        res.append(-heap[0][0])

    return res

arr = [1,3,-1,-3,5,3,6,7]
k = 3

sliding_window_maximum(arr, k)


# Dry Run

# arr = [1,3,-1,-3,5,3,6,7]
# k = 3

# iteration 1:
# heappush(heap, (-1,0))
# heappush(heap, (-3,1))
# heappush(heap, (1,2))

# heap = [(-3,1),(-1,0),(1,2)]

# res = [3]

# iteration 2:
# for i = 3:
# while heap and heap[0][1] <= 3 - 3 => while heap and 0 <= 0 => while heap and True
# heappop(heap) => heappop(heap) => heap = [(-1,0),(1,2)]

# heappush(heap, (-3,3)) => heappush(heap, (-3,3))

# heap = [(-3,3),(-1,0),(1,2)]

# res = [3,3]

# iteration 3:
# for i = 4:
# while heap and heap[0][1] <= 4 - 3 => while heap and 0 <= 1 => while heap and False
# heappush(heap, (-5,4)) => heappush(heap, (-5,4))

# heap = [(-5,4),(-3,3),(-1,0),(1,2)]

# res = [3,3,5]

# iteration 4:
# for i = 5:
# while heap and heap[0][1] <= 5 - 3 => while heap and 3 <= 2 => while heap and False
# heappush(heap, (-3,5)) => heappush(heap, (-3,5))

# heap = [(-5,4),(-3,5),(-3,3) ,(-1,0),(1,2)]

# res = [3,3,5,5]

# iteration 5:
# for i = 6:
# while heap and heap[0][1] <= 6 - 3 => while heap and 5 <= 3 => while heap and False
# heappush(heap, (-6,6)) => heappush(heap, (-6,6))

# heap = [(-6,6),(-5,4),(-3,5),(-3,3) ,(-1,0),(1,2)]

# res = [3,3,5,5,6]

# iteration 6:
# for i = 7:
# while heap and heap[0][1] <= 7 - 3 => while heap and 6 <= 4 => while heap and False
# heappush(heap, (-7,7)) => heappush(heap, (-7,7))

# heap = [(-7,7),(-6,6),(-5,4),(-3,5),(-3,3) ,(-1,0),(1,2)]

# res = [3,3,5,5,6,7]



```

```python
# LRU Cache:

from collections import OrderedDict

class LRUCache:

    def __init__(self, capacity):
        self.cache = OrderedDict()
        self.capacity = capacity
    
    def get(self, key):
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)
        return self.cache[key]
    
    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
            self.cache.popitem(last = False)

# test the LRU Cache:
            
cache = LRUCache(2)
cache.put(1, 1)  # cache is {1=1}
cache.put(2, 2) # cache is {1=1, 2=2}
print(cache.get(1)) # return 1
cache.put(3, 3) # LRU key was 2, evicts key 2, cache is {1=1, 3=3}
print(cache.get(1)) # return 1
cache.put(4, 4) # LRU key was 2, evicts key 2, cache is {1=1, 4=4}
print(cache.get(4)) # return 1
```

```python
# Is Identical -- subtree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    
    def isIdentical(self,root1, root2):
        if root1 is None and root2 is None:
            return True
        if root1 is None or root2 is None or root1.data != root2.data:
            return False

        if not self.isIdentical(root1.left, root2.left):
            return False
        
        if not self.isIdentical(root1.right, root2.right):
            return False
        
        return True
    
    def isSubtree(self,T, S):
        if S is None:
            return True
        
        if T is None:
            return False
        
        if self.isIdentical(T, S):
            return True 
        
        if self.isSubtree(T.left, S):
            return True
        
        if self.isSubtree(T.right, S):
            return True
        
        return False
    
root1 = Node(10)
root1.left = Node(8)
root1.right = Node(2)
root1.left.left = Node(3)
root1.left.right = Node(5)

root2 = Node(8)
root2.left = Node(3)
root2.right = Node(3)


print(root1.isSubtree(root1, root2))


# TC: O(n^2)
```

```python
# Two sum:

def two_sum(arr,target):
    mapp = {}

    for i in range(len(arr)):
        x = arr[i]
        if target - x in mapp:
            return [mapp[target-x]+1, i+1]
        mapp[x] = i

print(two_sum([2,7,11,15], 9))
```

```python
class TwoSm:
    def __init__(self) -> None:
        self.mapp = {}
    
    def add(self, x):
        if x in self.mapp:
            self.mapp[x] += 1
        else:
            self.mapp[x] = 1

    def find(self, target):
        for key in self.mapp:
            if target - key in self.mapp:
                return True
        return False

two_sum = TwoSm()
two_sum.add(1)
two_sum.add(2)
two_sum.add(3)
print(two_sum.find(4))
                
```

```python
# Amazon has multiple delivery centers and delivery warehouses all over  the world! The world is represented by number line from -10^9 to 10^9. There are n delivery centers, the ith one at the location center[i]. A location x is called a suitable location for a warehouse if it is possible to bring all the products to that point by travelling a distance of no more than d. At any one time, products can be brought from one delivery center and placed at point x. Given the positions of n delivery centers, calculate the number of suitablelocations in the world. That is, calculate the number of points x on the number line (-10^9 <= x <= 10^9) where the travel distance required to bring all the products to that point is less than or equal to d.

# Note: The distance between point x and center[i] is | x- center[i]|, their absolute difference.

# Example:
#  n = 3
#  center = [-2,1,0]
#  d = 8
#  output = 3

# Explanation: The variuos locations along with the distance traveled to bring all the treasures at the point are-
#  Locate warehouse at x = -3. First bring the products from center[0] = -2 covering a distance of |-3 + (-2)| = 1 to reach the center and |-3 - (-2)| = 1 to return. Similarly we bring products from centers 1 and 2 to point -3 for a total distance of 1 + 1 + 4 + 4 + 3 + 3 = 16

# Locate warehouse at x = 0. The total distance traveled is 2 * | 0 - (-2)| + 2 * | 0 - 1| + 2 * | 0 - 0| = 4 + 2 = 6
# Locate warehouse at x = -1 The total distance traveled is 2 * | -1 - (-2)| + 2 * | -1 - 1| + 2 * | -1 - 0| = 2 + 4 + 2 = 8
# Locate warehouse at x = 1 The total distance traveled is 2 * | 1 - (-2)| + 2 * | 1 - 1| + 2 * | 1 - 0| = 6 + 0 + 2 = 8

# There are 3 locations where the total distance traveled is less than or equal to 8

# Example 2:
# n = 4
# center = [2,0,3,-4]
# d = 22
# output = 5

# Explanation: The various locations along with the distance traveled to bring all the treasures at the point are:
# Locate warehouse at x = -1 The total distance traveled is 2 * | -1 - 2| + 2 * | -1 - 0| + 2 * | -1 - 3| + 2 * | -1 - (-4)| = 6 + 2 + 8 + 6 = 22
# Locate warehouse at x = 0 The total distance traveled is 2 * | 0 - 2| + 2 * | 0 - 0| + 2 * | 0 - 3| + 2 * | 0 - (-4)| = 4 + 0 + 6 + 8 = 18
# Locate warehouse at x = 1 The total distance traveled is 2 * | 1 - 2| + 2 * | 1 - 0| + 2 * | 1 - 3| + 2 * | 1 - (-4)| = 2 + 2 + 4 + 10 = 18
# Locate warehouse at x = 2 The total distance traveled is 2 * | 2 - 2| + 2 * | 2 - 0| + 2 * | 2 - 3| + 2 * | 2 - (-4)| = 0 + 4 + 2 + 12 = 18
# Locate warehouse at x = 3 The total distance traveled is 2 * | 3 - 2| + 2 * | 3 - 0| + 2 * | 3 - 3| + 2 * | 3 - (-4)| = 2 + 6 + 0 + 14 = 22

# There are 5 locations where the total distance traveled is less than or equal to 22

def suitableWareHouses(points,d):
    # Check if a given point is suitable
    def feasible(mid):
        total = 0
        for point in points:
            total+=abs(mid - point) * 2
        return total <=d
    def binarySearchRight(start,end):
        low = start
        high = end
        while low <= high:
            print(low, high)
            mid = low + (high - low)//2
            if feasible(mid):
                print("if", mid)
                low = mid + 1
            else:
                print("else", mid)
                high = mid - 1
        return low - 1
    def binarySearchLeft(start,end):
        low = start
        high = end
        while low <= high:
            mid = low + (high - low)//2
            if feasible(mid):
                high = mid - 1
            else:
                low = mid + 1
        return low
    suitablePoints = []
    # Find the rightmost suitable warehouse point
    
    rightLmit = binarySearchRight(0,max(points)+d)
    # Find the leftmost suitable warehouse point
    leftLimit = binarySearchLeft(min(points) - d,-1)
    # All the integer points between leftmost and rightmost points are also suitable
    for i in range(leftLimit,rightLmit+1):
        suitablePoints.append(i)
    return (suitablePoints)

# Time complexity: O(nlogn)
# Space complexity: O(1)

# Test case 1
n = 4
center = [2,0,3,-4]
d = 22
print(suitableWareHouses(center, d)) # 5
```

```python
# Given an array nums of n integers where n > 1, return an array output such that output[i] is equal to the product of all the elements of nums except nums[i].

# Example: nums = [1,2,3,4]
# output = [24,12,8,6]

# Explanation: The product of all the elements except nums[0] is 2*3*4 = 24, the product of all the elements except nums[1] is 1*3*4 = 12, the product of all the elements except nums[2] is 1*2*4 = 8, the product of all the elements except nums[3] is 1*2*3 = 6

# left_products = [1,1,2,6]
# right_products = [24,12,4,1]

def product_except_self(nums):
    n = len(nums)
    left_products = [1] * n  # [1,1,1,1]
    right_products = [1] * n # [1,1,1,1]
    output = [1] * n # [1,1,1,1]

    for i in range(1, n):
        left_products[i] = left_products[i-1] * nums[i-1]  # [1,1,2,6]

    for i in range(n-2, -1, -1):
        right_products[i] = right_products[i+1] * nums[i+1]

    for i in range(n):
        output[i] = left_products[i] * right_products[i]

    return output

# Time complexity: O(n)
# Space complexity: O(n)

#Follow up: Could you solve it with constant space complexity? (The output array does not count as extra space for the purpose of space complexity analysis.)

def product_except_self_constant_space(nums):
    n = len(nums)
    output = [1] * n # [1,1,1,1]

    for i in range(1, n):
        output[i] = output[i-1] * nums[i-1] # [1,1,2,6]

    right_product = 1
    for i in range(n-1, -1, -1):
        output[i] = output[i] * right_product # [24,12,8,6]
        right_product *= nums[i] 

    return output


# Time complexity: O(n)
# Space complexity: O(1)
```

```python
# Given a non-empty string s, you may delete at most one character. Judge whether you can make it a palindrome.
def valid_palindrome(s):
    def is_palindrome_range(i, j):
        return all(s[k] == s[j - k + i] for k in range(i, j))

    for i in range(len(s) // 2):
        if s[i] != s[~i]:
            j = len(s) - 1 - i
            return is_palindrome_range(i + 1, j) or is_palindrome_range(i, j - 1)
    return True

# Test cases
print(valid_palindrome("abca"))  # True
print(valid_palindrome("racecar"))  # True
print(valid_palindrome("abcdef"))  # False
```
