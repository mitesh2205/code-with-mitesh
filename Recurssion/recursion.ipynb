## Recurssion and Backtracking

```python
# non ending recursion:

def infinite_recursion():
    print(1)
    infinite_recursion()

infinite_recursion()


#  stack overflow error occurs when the recursion depth exceeds the maximum recursion depth

#  to avoid the stack overflow error we have to use the base condition to stop the recursion
```

```python
def print_linearly_1_n(i,n):
    if i>n: return
    print(i)
    print_linearly_1_n(i+1,n)

print_linearly_1_n(1,4)

#  reverse
print("reverse")
def print_reverse(n):
    if n<=0:
        return
    print(n)
    print_reverse(n-1)

print_reverse(4)
```

```python
#  Parameterised way to do sum of the numbers from 1 to n

def sum_parameterized(i,sum):
    if i<0: return
    # print(i)
    print(sum)
    sum_parameterized(i-1, sum+i)


sum_parameterized(3,0)

# functional

print("Functional")

def sum_functional(n):
    if n==0:
        return 0
    return n + sum_functional(n-1)

print(sum_functional(3))
```

```python
# Factorial 

def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(6))


# Recurssion tree for the above problem is :

# f(n) = n * f(n-1) 
# f(n-1) = n-1 * f(n-2)
# f(n-2) = n-2 * f(n-3)
# f(n-3) = n-3 * f(n-4)
```

```python
# Fibonacci number of a Nth:

def fibonacci(n):
    if n == 0 or n == 1:
        return n
    
    fn1 = fibonacci(n-1)
    fn2 = fibonacci(n-2)

    return fn1 + fn2

print(fibonacci(4))

# Print the Fibonacci Series till the nth number:

def fibonacci_series(n):
    t1 = 0
    t2 = 1

    series = str(t1) + " " + str(t2) + " "
    
    for i in range(2,n):
        next_term = t1 + t2
        t1 = t2
        t2 = next_term
        series += str(next_term) + " "
    return series
        

print(fibonacci_series(8))
        
```

```python
# Check if array is sorted or not

def is_sorted(arr,i):
    if i == len(arr)-1:
        return True
    if arr[i] > arr[i+1]:
        return False
    return is_sorted(arr,i+1)

print(is_sorted([1,2,3,4,5],0))
 
```

```python
# FirstOccurance in the array:

def first_occurance(arr, i, key):
    if i == len(arr):
        return -1
    
    if arr[i] == key:
        return i
    
    return first_occurance(arr, i+1, key)

print(first_occurance([1,2,3,4,5,6,7,8,9],0,5))
```

```python
# LastOccurance in the array for the given value:

def last_occurance(arr, i, key):
    if i == -1:
        return -1
    
    if arr[i] == key:
        return i
    
    return last_occurance(arr, i-1, key)

print(last_occurance([1,2,3,4,5,5,6,7,8,9],9,5))
```

```python
# Find X to the power of N:

def power_function(x,n):
    if n == 0:
        return 1
    return x * power_function(x,n-1)

# print(power_function(2,3))


def power_optimized(x,n):
    if n == 0:
        return 1
    
    half_power = power_optimized(x,n//2)
    half_power_square = half_power * half_power

    if n % 2 == 0:
        return half_power_square
    else:
        return x * half_power_square
    
print(power_optimized(5,3))
```

```python
# Tiling Problem:

def tiling_problem(n):
    if n == 0 or n == 1:
        return 1

    return tiling_problem(n-1) + tiling_problem(n-2)

print(tiling_problem(4))
```

```python
# Remove duplicates from the string

def remove_duplicates(s,i, mapp, new_s):

    if i == len(s):
        return new_s
    
    if mapp[ord(s[i]) - ord('a')] == False:
        new_s += s[i]
        mapp[ord(s[i]) - ord('a')] = True
    
    return remove_duplicates(s,i+1,mapp,new_s)

print(remove_duplicates("abbccdeff", 0, [False] * 26, ""))
```

```python
# For a given Integer array of size N. You have to find all the occurances (Indices) of the given element (key) and pring them

def all_occurance(arr, key, i):
    if i == len(arr):
        return
    
    if arr[i] == key:
        print(i)
    all_occurance(arr, key, i + 1)

print(all_occurance([3,2,4,5,6,2,7,2,2], 2,0))
```

```python
# Convert the number in to the string, EG 2019 -> two zero one nine

def convert_to_string(n, mapp):
    if n == 0:
        return
    
    convert_to_string(n//10, mapp)
    print(mapp[n%10], end=" ")

mapp = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]

convert_to_string(2019, mapp)
```

```python
# Find length of string using the recurssion:

def len_of_string(s):
    if s == "":
        return 0
    return 1 + len_of_string(s[1:])

print(len_of_string("abcdefg"))
```

```python
# Print the Contiguous substring that starts and ends with the same letter

def count_substring_same_start_end(s,i,j,n):
    if n == 1:
        return 1
    if n <= 0:
        return 0

    res = count_substring_same_start_end(s, i+1,j,n-1) + count_substring_same_start_end(s,i,j-1,n-1) - count_substring_same_start_end(s,i+1,j-1,n-2)
    
    if s[i] == s[j]:
        res += 1
    
    return res

print(count_substring_same_start_end("aba", 0, 2, 3))


# DP for the above problem

def count_substring_same_start_end_dp(s):
    n = len(s)
    dp = [[0 for i in range(n)] for j in range(n)]
    print(dp)
    count = 0

    for i in range(n):
        dp[i][i] = 1
        count += 1

#   this below loop is for the two characters in the string that are same and are at the same index 
    for i in range(n-1):
        if s[i] == s[i+1]:
            dp[i][i+1] = 1
            count += 1

    for i in range(n-2, -1, -1):  # this loop is for the characters that are at the different index
        for j in range(i+2, n):
            if s[i] == s[j] and dp[i+1][j-1]:
                dp[i][j] = 1
                count += 1

    return count


print(count_substring_same_start_end_dp("aba"))
```

```python
# Longest Increasing Subsequence 

def longest_increasing_subsequence(arr, i, n):
    if i == n:
        return 0
    
    res = 1
    for j in range(i+1, n):
        if arr[j] > arr[i]:
            res = max(res, 1 + longest_increasing_subsequence(arr, j, n))
    
    return res

print(longest_increasing_subsequence([10,22,9,33,21,50,41,60,80], 0, 9))

# DP for the above problem

def longest_increasing_subsequence_dp(arr):
    n = len(arr)
    dp = [1] * n

    for i in range(1,n):
        for j in range(i):
            if arr[i] > arr[j]:
                dp[i] = max(dp[i], 1 + dp[j])

    print(dp)
    return max(dp)

print(longest_increasing_subsequence_dp([10,22,9,33,21,50,41,60,80]))
```
